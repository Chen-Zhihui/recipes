```python
#pyfaidx
from pyfaidx import Fasta

genes = Fasta('./capp-seq/selector-clean.fa', key_function = lambda x: x.split('=')[1])
genes
```

```python

for g in genes.keys() :
    ch, rng = g.split(":")
    ch = ch.split("r")[1]
    start, end = rng.split("-")
    print("{}=[{},{}], len={}".format(ch, start, end, len(genes[g])-1000))
```
